```python
import sklearn.linear_model
import sklearn.datasets
import sklearn.svm
import sklearn.metrics
import sklearn.decomposition
import sklearn.feature_extraction.text
import sklearn.utils.sparsefuncs
```

```python
import re
input_corpus=[]
i=0
for rows in data['product desc']:
#     print rows,type(rows)
#     print i
#     print re.sub(r'[^\x00-\x7F]',' ', rows)
#     i=i+1
    input_corpus.append(re.sub(r'[^\x00-\x7F]',' ', rows))

    
print type(input_corpus)
```

```python
vectorizer = sklearn.feature_extraction.text.CountVectorizer(
    stop_words='english',
    ngram_range=(1, 1),  #ngram_range=(1, 1) is the default
    dtype='double',
)
data = vectorizer.fit_transform(input_corpus)
```

```python
from wordcloud import WordCloud, STOPWORDS
stopwords = set(STOPWORDS)
wordcloud = WordCloud(
                          background_color='white',
                          stopwords={'slits','sleeves','has','a','and','hem','neck',"hem'",'kurta'},
#                           stopwords=stopwords,
    
                          max_words=200,
                          max_font_size=40, 
                          random_state=42
                         ).generate(str(input_corpus))
```
